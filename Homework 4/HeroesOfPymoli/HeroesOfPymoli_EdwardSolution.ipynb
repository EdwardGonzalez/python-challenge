```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
PrData = pd.read_csv(file_to_load)

PrData.head()
```

## Player Count

* Display the total number of players


```python
#Solution
TNumPlayers = PrData.SN.nunique()
print("The total numbers of players is: "+str(TNumPlayers))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Solutions
#Get Total unique items
UniqueItems = PrData['Item ID'].nunique()

#Get Avg.prive
AvgPrice = PrData.Price.mean()

#Number of Prucchases
TotalPr = PrData['Purchase ID'].nunique()


#Total Revenue
TotalRev = PrData.Price.sum()
```

```python
#Generate DataFrame
Ex2 = {
        'Number of Unique Items': [UniqueItems],
        'Average Price': [AvgPrice],
        'Number of Purchases': [TotalPr],
        'Total Revenue':[TotalRev]
      }
pdEx2=pd.DataFrame(Ex2)

#colocar Formato
pdEx2['Average Price'] = pdEx2['Average Price'].map('${:,.2f}'.format)
pdEx2['Total Revenue'] = pdEx2['Total Revenue'].map('${:,.2f}'.format)

pdEx2
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Solution
Ex3 = pd.DataFrame(PrData.groupby('Gender')['SN'].nunique())

Ex3['Percentage of Players'] = Ex3['SN']/Ex3['SN'].sum()*100
Ex3['Percentage of Players'] = Ex3['Percentage of Players'].map('{:,.2f}%'.format)
Ex3.rename(columns={"SN":"Total Count"})
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Solutions
#Get Purchase Count
Ex4 = pd.DataFrame(PrData.groupby('Gender')['Gender'].count())
Ex4['Average Purchase Price'] = PrData.groupby('Gender')['Price'].mean()
Ex4['Total Purchase Value'] = PrData.groupby('Gender')['Price'].sum()
Ex4['Avg Total Purchase per Person'] = PrData.groupby('Gender')['Price'].sum()/PrData.groupby('Gender')['SN'].nunique()

#Colocar Formato
Ex4['Average Purchase Price'] = Ex4['Average Purchase Price'].map('${:,.2f}'.format)
Ex4['Total Purchase Value'] = Ex4['Total Purchase Value'].map('${:,.2f}'.format)
Ex4['Avg Total Purchase per Person'] = Ex4['Avg Total Purchase per Person'].map('${:,.2f}'.format)

Ex4.rename(columns={"Gender":"Purchase Count"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
 #solution
Ex5 = PrData
Ex5['Categoria Edad'] = pd.cut(np.array(PrData['Age']),[0,10,15,20,25,30,35,40,100],right=False, labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','>40'])

S = pd.DataFrame(Ex5.groupby('Categoria Edad')['SN'].nunique())
S['Percentage of Player']=  S['SN']/S['SN'].sum()*100
S['Percentage of Player'] = S['Percentage of Player'].map('{:,.2f}%'.format)
S.rename(columns={"SN":"Total Count"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Ex8 = pd.DataFrame(PrData.groupby(['Item ID','Item Name'])['Gender'].count())
Ex8['Price']=PrData.groupby(['Item ID','Item Name'])['Price'].mean()
Ex8['Total Purchase Value']=PrData.groupby(['Item ID','Item Name'])['Price'].sum()
Ex8['Price'] = Ex8['Price'].map('${:,.2f}'.format)
Ex8['Total Purchase Value'] = Ex8['Total Purchase Value'].map('${:,.2f}'.format)
Ex8.rename(columns={"Gender":"Purchase Count"})

MPOPI =Ex8.sort_values(['Gender'],ascending=False)
MPOPI.rename(columns={"Gender":"Purchase Count"}).head()
```
